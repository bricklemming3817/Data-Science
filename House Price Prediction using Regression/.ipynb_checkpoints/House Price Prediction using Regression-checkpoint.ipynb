Creating functions to encoder non-numeric datas to numeric data

```python
def encoder(parameter):
    num_data = len(parameter.unique())
    index_data = [x for x in range(num_data)]
    data = list(parameter.unique())
    data_map = []

    for i in zip(data, index_data):
        data_map.append(i)
    data_map = dict(data_map)
    print(data_map)
    return(data_map)

def labeler(data):
    n = data.shape[0]
    columns = data.columns
    for cols in columns:
        for i in range(0, n):
            if type(data[cols][i]) == str:
                label = encoder(data[cols])
                data[cols] = data[cols].map(label)
    return(data)
    
    
    
```

```python
train_enc = labeler(train_data)
test_enc = labeler(test_data)
```

## Model Building

Converting the data sets to arrays and normalising the values

```python
x_train = np.asarray(train_enc.drop(['SalePrice'], axis = 1))
y_train = np.asarray(train_enc['SalePrice'])
x_test = np.asarray(test_enc)
```

```python
x_train = StandardScaler().fit(x_train).transform(x_train)
x_test = StandardScaler().fit(x_test).transform(x_test)
```

To evaluate models, we are splitting training dataset to different subesets

```python
x_mod_train, x_mod_test, y_mod_train, y_mod_test = train_test_split(x_train, y_train, test_size = 0.3)
```

#### Model I: LineraRegression
For this model we need to determine the degree of the polynomial which need to be fit 

```python
lr = LinearRegression()
```

```python
rsqt_test = []
order = [x for x in range(1,6)]

for n in order:
    pr = PolynomialFeatures(degree = n)
    x_train_pr = pr.fit_transform(x_mod_train)
    x_test_pr = pr.fit_transform(x_mod_test)
    
    lr.fit(x_train_pr, y_mod_train)
    rsqt_test.append(lr.score(x_test_pr, y_mod_test))
    print('n = {}, R^2 = {}'.format(n, lr.score(x_test_pr, y_mod_test)))
plt.plot(order, rsqt_test)
plt.xlabel('order n')
plt.ylabel('R^2')
plt.show()
                     
    
    
```

From the plot we can see the R^2 values is greatest for n = 1. 
Building linear regression model for first degree

```python
order = np.array(rsqt_test).argmax()+1
pr = PolynomialFeatures(degree = order)
x_train_pr = pr.fit_transform(x_mod_train)
x_test_pr = pr.fit_transform(x_mod_test)
    
lr.fit(x_train_pr, y_mod_train)
y = lr.predict(x_test_pr)

sns.distplot(y_mod_test, hist = False, label = 'y_test data')
sns.distplot(y, hist = False, label = 'predicted y values')
```

#### Model II: Ridge Regressor

```python
parameters= [{'alpha': [0.001,0.1,1, 10, 100, 1000, 10000, 100000, 100000]}]
rr=Ridge()
grid = GridSearchCV(rr, parameters,cv=4)
```

```python
grid.fit(x_mod_train, y_mod_train)
best_model = grid.best_estimator_
print('R^2 score: ', best_model.score(x_mod_test, y_mod_test))
best_model
```

```python
yh = best_model.predict(x_mod_test)
sns.distplot(y_mod_test, hist = False, label = 'y_test data')
sns.distplot(yh, hist = False, label = 'predicted y values')
```

#### Model III : Gradient Booster Regressor

```python
xgbr = xgb.XGBRegressor()
```

```python
xgbr.fit(x_mod_train, y_mod_train)
cvs = cross_val_score(xgbr, x_mod_train, y_mod_train, cv = 10)
print('R^2 score: ', xgbr.score(x_mod_test, y_mod_test))
print('model accuracy according to cross validation score: {} %'.format(round(cvs.mean()*100, 2)))
```

```python
y_hat = xgbr.predict(x_mod_test)
print('mean squared error: ', mean_squared_error(y_mod_test, y_hat ))
fig, axs = plt.subplots(figsize = (10,10))
sns.distplot(y_mod_test, hist = False, label = 'y_test data')
sns.distplot(y_hat, hist = False, label = 'predicted y values')
plt.show()
```

From the above three models we can see that XGB Regressor we the best perform model with the maximum accuracy.
So we chose XGB Regressor for final prediction
